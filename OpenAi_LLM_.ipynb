```python
import random
import time
# Example Customer Data
customer_profile = {
    "age": 22,
    "gender": "Male",
    "marital_status": "Single",
    "education": "Bachelor",
    "occupation": "Software Engineer",
    "monthly_income": 5000,
    "product_price": 1500,
    "preferred_repayment": "Pay in 4, Monthly Installments"
}

# Example BNPL Provider Data
bnpl_providers = [
    {
        "Provider": "Affirm",
        "Interest_Rate": 15,
        "Transaction_Limit": 5000,
        "Max_Credit_Limit": 20000,
        "Repayment_Options": "Pay in 4, Monthly Installments"
    },
    {
        "Provider": "Afterpay",
        "Interest_Rate": 0,
        "Transaction_Limit": 3000,
        "Max_Credit_Limit": 4000,
        "Repayment_Options": "Pay in 4"
    },
    {
        "Provider": "Klarna",
        "Interest_Rate": 12,
        "Transaction_Limit": 4000,
        "Max_Credit_Limit": 5000,
        "Repayment_Options": "Pay in 3, Monthly Installments, Pay in 30 Days"
    },

]
transaction_history = [
    {
        "transaction_ID": f"T{random.randint(10000, 99999)}",
        "amount": random.randint(50, 1000),
        "time_of_transaction": time.time(),  # Current timestamp
        "user_id": "U12345"
    }
    for _ in range(5)  # Generate 5 random transactions
]
```

```python
import openai
import os

api_key = os.getenv("OPENAI_API_KEY")

client = openai.OpenAI(api_key=os.getenv(""))# replace with your open AI api key


def rank_bnpl_providers(customer_profile, bnpl_providers, transaction_history):
    # Format transaction history for the prompt
    transaction_text = "\n".join([
        f"- Transaction ID: {txn['transaction_ID']}, Amount: ${txn['amount']}, Time: {txn['time_of_transaction']}, User ID: {txn['user_id']}"
        for txn in transaction_history
    ])

    # Create a prompt for OpenAI LLM
    prompt = f"""
    You are an expert AI assistant who evaluates and ranks BNPL (Buy Now, Pay Later) providers based on user financial details, past transaction history, provider terms, and approval likelihood.

    ### User Details:
    - Age: {customer_profile['age']}
    - Gender: {customer_profile['gender']}
    - Marital Status: {customer_profile['marital_status']}
    - Education: {customer_profile['education']}
    - Occupation: {customer_profile['occupation']}
    - Monthly Income: ${customer_profile['monthly_income']}
    - Product Price: ${customer_profile['product_price']}
    - Preferred Repayment: {customer_profile['preferred_repayment']}

    ### Transaction History:
    {transaction_text}

    ### BNPL Providers:
    {bnpl_providers}

    ### Provider Terms Highlights:

    **Klarna:**
    - **Cashback Program:** Cashback is earned on qualifying purchases (excluding taxes and fees) and credited to the Klarna balance (rounded down to the nearest cent).
    - **Eligibility & Approval:** Available only to customers residing in Ireland with a valid billing address; purchases require approval from participating stores.
    - **Terms Flexibility:** Cashback rates and eligibility may change, with limited liability under Irish law.

    **Affirm:**
    - **Eligibility & User Requirements:** Users must be at least 18, reside in the UK, have a UK mobile number, and meet strict creditworthiness and affordability criteria.
    - **Service Model:** Affirm pays the merchant upfront while the user repays via a loan governed by a separate credit agreement.
    - **Data & Decision Making:** Uses personal data for identity verification, performs automated credit checks, and reports payment behavior to credit agencies.
    - **Liability & Updates:** Liability is limited, and terms can change with advance notice. Users are responsible for accurate account information and security.

    **Clearpay:**
    - **Account & Eligibility:** Available to UK residents (excluding Channel Islands), aged 18+ with a valid UK billing address, UK mobile number, and payment method.
    - **Plan Structure:** Provides a fixed 4-installment, interest-free plan with payments automatically deducted every two weeks after pre-authorization checks.
    - **Late Fees & Penalties:** Charges a £6 late fee for missed installments (with orders of £24 or more possibly incurring two fees capped at the lower of £24 or 25% of the purchase price).
    - **Refunds & Cancellations:** Adjusts payment schedules for refunds; orders may be canceled before delivery if issues arise.
    - **Security & Account Management:** Emphasizes secure account management, with measures to suspend or close accounts for suspicious activities.

    #### Evaluation Criteria:
    1. **Approval Likelihood:** Based on the user's income, past transaction patterns, and provider approval conditions.
    2. **Interest Rate & Loan/Plan Terms:** Lower interest rates or favorable installment terms are preferred.
    3. **Repayment Options Match:** Does the provider offer the user’s preferred repayment method?
    4. **Financing Capacity:** Can the provider finance the purchase based on past spending patterns and BNPL transaction limits?
    5. **Clarity & Favorability of Terms:** Consider how clearly and favorably each provider presents its eligibility requirements, fees, and additional benefits.

    #### Instructions:
    Rank the BNPL providers **from best to worst** using the evaluation criteria above. Provide a numbered list that includes only the provider names (as given in the BNPL Providers section) along with a brief explanation for each ranking.

    Example Output:
    1. Provider A – Explanation
    2. Provider B – Explanation
    3. Provider C – Explanation

    Ensure your final answer strictly follows this format without any extra commentary.
    """

    # Call OpenAI API with the new format
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a financial AI assistant."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content

# Run the function with mock data
ranking_result = rank_bnpl_providers(customer_profile, bnpl_providers, transaction_history)
print(ranking_result)

```

```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

# Choose the free and open model
model_name = "EleutherAI/gpt-neo-125M"

# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Create the text generation pipeline without the use_auth_token parameter
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Sample customer profile
customer_profile = {
    'age': 30,
    'gender': 'Male',
    'marital_status': 'Single',
    'education': "Bachelor's Degree",
    'occupation': 'Software Engineer',
    'monthly_income': 5000,
    'product_price': 1200,
    'preferred_repayment': 'Monthly Installments'
}

# Sample BNPL provider details
bnpl_providers = """
Provider A:
  - Approval criteria: High income thresholds, low interest rate.
  - Transaction limits: Up to $1500.
  - Repayment: Monthly installments.

Provider B:
  - Approval criteria: Moderate income thresholds, moderate interest rate.
  - Transaction limits: Up to $1000.
  - Repayment: Bi-weekly installments.

Provider C:
  - Approval criteria: Flexible income thresholds, higher interest rate.
  - Transaction limits: Up to $2000.
  - Repayment: Flexible repayment options.
"""

# Create the prompt for ranking BNPL providers
#refining the prompts
prompt = f"""
You are an AI assistant that ranks BNPL (Buy Now, Pay Later) providers based on user financial details, BNPL provider terms, and approval likelihood.

### User Details:
- Age: {customer_profile['age']}
- Gender: {customer_profile['gender']}
- Marital Status: {customer_profile['marital_status']}
- Education: {customer_profile['education']}
- Occupation: {customer_profile['occupation']}
- Monthly Income: ${customer_profile['monthly_income']}
- Product Price: ${customer_profile['product_price']}
- Preferred Repayment: {customer_profile['preferred_repayment']}

### BNPL Providers:
{bnpl_providers}

#### Instructions:
Based on the following criteria:
- **Approval Likelihood:** Can the user get approved based on income, price, and provider limits?
- **Interest Rate:** Lower is better.
- **Repayment Options Match:** Does the provider offer the user's preferred repayment method?
- **Transaction Limit & Max Credit Limit:** Can the provider finance this purchase?

Please rank the BNPL providers **from best to worst**. Your answer should be a numbered list that includes **only the provider names as they appear in the BNPL Providers section** along with a brief explanation for each ranking. Do not include any extra headers or generic labels like "Lowest" or "Highest." For example:

1. Provider A – Explanation of why Provider A ranks first.
2. Provider B – Explanation of why Provider B ranks second.
3. Provider C – Explanation of why Provider C ranks third.

Ensure your final answer follows this format.
"""


# Generate and print the response, adjusting generation parameters for more output
response = generator(
    prompt,
    max_length=len(tokenizer.encode(prompt)) + 100,  # generate 100 new tokens beyond the prompt
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    top_k=50
)

print(response[0]["generated_text"])
```
